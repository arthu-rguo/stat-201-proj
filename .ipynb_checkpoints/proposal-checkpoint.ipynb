# Examining the difference between mean minimum and maximum daily temperatures at UBC and SFU

## 1. Introduction

Students choosing between the University of British Columbia (UBC) and Simon Fraser University (SFU) may consider campus weather in their decision-making process. Studies have shown that temperature has an impact on student productivity and emotional stability (1, 2). Temperature has also been observed to impact the daily academic routines of students—colder temperatures tend to lead to lower class attendance, while hotter temperatures result in decreased self-study hours (3).

Our study aims to see whether there is a significant difference between the mean minimum and maximum daily temperatures at UBC and SFU. The dataset we will use compiles daily weather variables recorded from January to June, 1995 (4). Our location parameters will be the true mean minimum and maximum daily temperatures across all years for UBC and SFU, which we will estimate using the bootstrap distributions generated from our dataset. Overall, we hope our findings will support future research on relationships between temperature and academic success. 

## 2. Preliminary Results

### 2.1. R Preamble

```r
# Run this first.
library(tidyverse)
library(tidymodels)

set.seed(5)
options(repr.plot.width = 6, repr.plot.height = 4)

SRC <-
    c("https://raw.githubusercontent.com/arthu-rguo/stat-201-proj/main/ubc.csv",
      "https://raw.githubusercontent.com/arthu-rguo/stat-201-proj/main/sfu.csv")
```

### 2.2. Loading and wrangling the dataset

Note that the raw UBC and SFU datasets span different date ranges. However, our analysis will focus mainly on the minimum and maximum daily temperatures, which might vary depending on the date. To prevent this from potentially skewing our results, we filter our wrangled dataset such that it spans the intersection of the UBC and SFU date ranges (from January 1 to June 30, 1995).

```r
# Load and label the raw datasets for UBC and SFU, respectively.
ubc_raw <- read_csv(SRC[1], show_col_types = FALSE) |>
    mutate(location = as.factor("UBC"))

sfu_raw <- read_csv(SRC[2], show_col_types = FALSE) |>
    mutate(location = as.factor("SFU"))

# Join the datasets, select the relevant variables, and tidy the data.
ubc_sfu_raw <- bind_rows(ubc_raw, sfu_raw) |>
    rename(date_time = "Date/Time", 
           min_temp_c = "Min Temp (°C)", max_temp_c = "Max Temp (°C)") |>
    select(location, date_time, min_temp_c, max_temp_c) |>
    filter(!is.na(min_temp_c), !is.na(max_temp_c))

# Find the intersection of the date ranges and filter the dataset.
date_range <- ubc_sfu_raw |>
    group_by(location) |>
    summarize(min = min(date_time), max = max(date_time)) |>
    ungroup() |>
    summarize(min = max(min), max = min(max))

ubc_sfu_data <- ubc_sfu_raw |>
    filter(between(date_time, date_range$min, date_range$max))

# Preview the dataset.
cat("Table 2.2.1. Preview of tidy data for UBC and SFU")
head(ubc_sfu_data)
tail(ubc_sfu_data)
```

### 2.3. Summary statistics

We calculate the summary statistics of interest for both UBC and SFU. The sample sizes are slightly different, but this should not a problem as both asymptotic and bootstrapping methods can account for this.

Overall, we can see that the mean minimum daily temperature at UBC in 1995 was 6.28°C, which is higher than the mean minimum daily temperature of 5.73°C at SFU, but both mean maximum daily temperatures are very similar at around 13.11°C.

```r
# Calculate means, standard deviations, and counts for both locations.
ubc_sfu_summary <- ubc_sfu_data |>
    group_by(location) |>
    summarize(mean_min = mean(min_temp_c), sd_min = sd(min_temp_c),
              mean_max = mean(max_temp_c), sd_max = sd(max_temp_c),
              n = n())

cat("Table 2.3.1. Summary statistics for UBC and SFU")
ubc_sfu_summary
```

### 2.4. Preliminary visualization

Plotting and comparing the sample distributions for both locations suggests that minimum and maximum daily temperatures might not be normally distributed—in particular, the distributions for UBC seem somewhat bimodal. Nonetheless, our sample sizes of 179 and 181 are still large enough for the Central Limit Theorem (CLT) to apply.

```r
# Plot sample distributions for min. daily temperatures.
ubc_sfu_min_sample_dist <- ubc_sfu_data |>
    ggplot(aes(x = min_temp_c, fill = location)) +
        geom_histogram(binwidth = 1, alpha = 0.8) +
        labs(title = "Figure 2.4.1. Sample distributions of min. daily temperatures",
             x = "Min. temperature (°C)", y = "Count", fill = "Location") +
        facet_wrap(~location, ncol = 1) +
        theme_bw() +
        theme(legend.position = "top", 
              strip.background = element_blank(), strip.text = element_blank())

# Plot sample distributions for max. daily temperatures.
ubc_sfu_max_sample_dist <- ubc_sfu_data |>
    ggplot(aes(x = max_temp_c, fill = location)) +
        geom_histogram(binwidth = 1, alpha = 0.8) +
        labs(title = "Figure 2.4.2. Sample distributions of max. daily temperatures",
             x = "Max. temperature (°C)", y = "Count", fill = "Location") +
        facet_wrap(~location, ncol = 1) +
        theme_bw() +
        theme(legend.position = "top", 
              strip.background = element_blank(), strip.text = element_blank())

ubc_sfu_min_sample_dist
ubc_sfu_max_sample_dist
```

To get a better idea of the centers and spreads of the sample distributions, we can also compare boxplots. We see that both sample distributions for minimum and maximum daily temperatures (especially SFU's maximum temperatures) are slightly right-skewed.

```r
# Create boxplots for min. daily temperatures.
ubc_sfu_min_boxplot <- ubc_sfu_data |>
    ggplot(aes(x = min_temp_c, fill = location)) +
        geom_boxplot(alpha = 0.8) + 
        labs(title = "Figure 2.4.3. Boxplots of min. daily temperatures",
             x = "Min. temperature (°C)", fill = "Location") +
        theme_bw() +
        theme(legend.position = "top", 
              axis.text.y = element_blank(), axis.ticks.y = element_blank())

# Create boxplots for max. daily temperatures.
ubc_sfu_max_boxplot <- ubc_sfu_data |>
    ggplot(aes(x = max_temp_c, fill = location)) +
        geom_boxplot(alpha = 0.8) + 
        labs(title = "Figure 2.4.4. Boxplots of max. daily temperatures",
             x = "Max. temperature (°C)", fill = "Location") +
        theme_bw() +
        theme(legend.position = "top", 
              axis.text.y = element_blank(), axis.ticks.y = element_blank())

ubc_sfu_min_boxplot
ubc_sfu_max_boxplot
```

## 3. Methods
The data used in this project was obtained from the Government of Canada (4). As a federal system, they must uphold their obligations to provide consistent and transparent data, providing a high degree of reliability. However, we will approach the data critically given that it is from 1995 and data collection has evolved over the years (5).

Our objective is to assess whether there is a significant difference between the mean minimum and maximum daily temperatures at UBC and SFU. Although we have computed summary statistics for our sample, single point estimates will not be not sufficient to draw conclusions about the population (Table 2.3.1). Instead, we will create bootstrap distributions to estimate the mean minimum and maximum temperatures at UBC and SFU across all years, which are our two population parameters of interest. We will then report two 95% confidence intervals which we anticipate will capture the true value of each population parameter.

Following this, we will conduct hypothesis tests. Our two null hypotheses ($H_0 : \mu_1 = \mu_2$) propose that there are no differences in the mean minimum and maximum daily temperature, while our alternative hypotheses ($H_A : \mu_1 \neq \mu_2$) state that there are. Despite the slight skew found in our preliminary results, our sample is sufficiently large to apply the CLT (Figures 2.4.1–2.4.4). Additionally, the data was randomly and independently sampled, meeting the assumptions of the CLT. Hence, we will use the CLT to determine p-values from our bootstrap distributions.  To conclude, we will set our significance level to 5% ($\alpha = 0.05$) and either reject or fail to reject our null hypotheses based on our calculated p-values, addressing our research objective.

We hope our findings will help guide students in making informed choices around university selection. Our research will contribute to the existing literature surrounding temperatures on university campuses, as well as the broader topic of relationships between temperature and academic success.
